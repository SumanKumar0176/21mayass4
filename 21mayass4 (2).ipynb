```python
13. What's the difference between tuples and lists?

Ans:

i.Tuples and lists are both sequential data types in Python, but they have some key differences:

ii.Mutability: Tuples are immutable, meaning once a tuple is created, its elements cannot be changed. In contrast, lists are mutable, 
allowing you to modify, add, or remove elements after the list is created.

Syntax: Tuples are defined using parentheses (), while lists are defined using square brackets []. For example:
my_tuple = (1, 2, 3)
my_list = [1, 2, 3]

iii.Usage: Tuples are typically used to represent a collection of related values that should not be changed, such as coordinates, database records, 
or fixed data. Lists, on the other hand, are used when you need a mutable collection that can be modified, expanded, or re-ordered as needed.

iv.Performance: Tuples are generally more efficient in terms of memory and performance compared to lists. Since tuples are immutable, Python can optimize
 their memory usage. Lists, being mutable, require more memory allocation and deallocation as they grow or shrink.

v.Methods: Lists have more built-in methods compared to tuples because of their mutability. Some list methods include append(), insert(), remove(), and 
sort(), which allow for modification and manipulation of the list. Tuples have fewer methods, primarily focusing on operations like indexing, slicing, and counting.

v.Iteration: Both tuples and lists can be iterated over using loops, but due to their immutability, tuples can provide a performance advantage in scenarios 
where the data does not need to be modified during iteration.

```
